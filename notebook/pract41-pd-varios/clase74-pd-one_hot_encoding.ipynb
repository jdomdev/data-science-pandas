Necesitamos datos numéricos en técnicas de ciencia de datos como el aprendizaje automático y los modelos de aprendizaje profundo. Comenzamos nuestro análisis con tipos de datos categóricos y numéricos. Al preparar los datos para el modelo, descartamos algunos tipos de datos categóricos si no los necesitamos, o usamos algunas técnicas como expresiones regulares y obtenemos valores numéricos. Me refiero a datos no numéricos como texto, objeto, fecha y hora, etc. con "datos categóricos" . Siempre habrá algunas columnas que necesitemos y no habrá valores numéricos que podamos obtener en expresiones regulares u otra función. Por lo tanto, para utilizar técnicas de ciencia de datos, necesitamos convertir estos datos no numéricos en datos numéricos.

En primer lugar, debemos decidir si estos datos cualitativos son nominales u ordinales. Este punto es importante, ya que definirá el estilo de nuestro enfoque. Si existe una jerarquía entre los datos, son ordinales. Si no existe jerarquía entre los datos, son nominales.

__* Datos cualitativos__

Tomemos como ejemplo los colores de los coches: no hay jerarquía, el coche puede ser negro, blanco o rojo. No se establece ningún orden entre los coches. Son datos nominales.

Ahora veamos los datos ordinales. Hay camisetas en la tienda y las tallas son grandes, medianas y pequeñas. Hay un rango. La grande es más grande que la mediana. Estos son datos ordinales.

#### Métodos de codificación

Existen algunos métodos de codificación, pero los más comunes y fáciles de implementar son la codificación de etiquetas y la codificación One Hot.

__* Codificación de etiquetas y codificación One Hot__

__1 - Codificación de etiquetas__

La codificación de etiquetas es más adecuada para datos ordinales, ya que asignamos números a cada valor único de los datos. Si utilizamos la codificación de etiquetas en datos nominales, le damos al modelo información incorrecta sobre nuestros datos. El algoritmo del modelo puede actuar como si existiera una jerarquía entre los datos.

__* Codificación de etiquetas para datos nominales__

Veamos el ejemplo. Hay una columna con tipos de puentes. Es nominal, ya que no hay una clasificación entre los tipos de puentes. En este ejemplo, se ha implementado la codificación de etiquetas y, con esta nueva columna, el algoritmo del modelo puede actuar como si hubiera una jerarquía entre los tipos de puentes.

__* Codificación de etiquetas para datos ordinales__

Este es el caso de uso correcto. Hay una columna de nivel de seguridad. Hay una jerarquía entre los datos, es ordinal. El nivel medio es inferior al nivel alto, por lo que puede tomar 2 para el nivel medio y 3 para el nivel alto. Podemos implementar la codificación de etiquetas y no será un problema para el algoritmo del modelo.

Existen distintas opciones para aplicar la codificación de etiquetas a los datos. Algunos ejemplos:

- Scikit-learn, la biblioteca de ciencia de datos más útil de Python, tiene métodos para esto.

__2 - Codificación One Hot__

La codificación one hot es la más común y la mejor opción, especialmente para datos nominales. También podemos usarla para datos ordinales. No importa si hay jerarquía o no.

Una codificación activa convierte la columna no numérica en columnas que contienen 1 y 0.

En este método, una columna no numérica se convierte en n * columnas. n se refiere a valores únicos en una columna no numérica.

__* Codificación One Hot__

En el ejemplo anterior, la columna de color tiene 3 valores únicos: rojo, azul y verde. Se convirtió en 3 columnas después de aplicar la codificación one hot. Una columna para cada valor: color_red, color_blue y color_green. Los identificadores de color azul son 2 y 4. En nuestro marco de datos convertido, observe las líneas 2 y 4; color_red y color_green son 0, pero color_blue es 1.

Puede ralentizar los modelos porque crea varias columnas. Más columnas significan menos rendimiento y más tiempo de entrenamiento. Es mejor usar codificación de etiquetas si tenemos datos ordinales y no los necesitamos.

Hay dos métodos para implementar una codificación en caliente;

- Scikit-learn tiene OneHotEncoder para aplicar una codificación one hot.

__from sklearn.preprocessing import OneHotEncoder__

- Podemos utilizar el método get_dummies de Pandas.

__* Codificación One Hot con Pandas get_dummies__

Podemos utilizar estos dos métodos, pero aplicar get_dummies es más fácil que Scikit-Learn OneHotEncoder y el método get_dummies tiene parámetros muy útiles. Hemos hablado del método get_dummies de Pandas en este enlace[https://hersanyagci.medium.com/encoding-with-pandas-get-dummies-2f95709adfd] .
